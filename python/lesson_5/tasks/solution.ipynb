```python
import pandas as pd

data = pd.read_csv('data.csv')
result = data[data['categories'].str.split(',').apply(len) > 20].groupby('province').agg({'longitude': 'min'})
result['longitude'] = result['longitude'].round(3)
result.to_csv('result.csv', sep=';')
```
